```python
con = sqlite3.connect('./chinook.db')
cur = con.cursor()
cur.execute("SELECT tracks.Name AS Song, genres.Name AS Genre, playlists.Name AS Play_List FROM tracks, genres, playlist_track, playlists,albums, artists WHERE tracks.GenreId=genres.GenreId AND playlists.PlaylistId=playlist_track.PlaylistId AND tracks.TrackId=playlist_track.TrackId AND tracks.AlbumId=albums.AlbumId AND albums.ArtistId=artists.ArtistId")
data = cur.fetchall()
df = pd.DataFrame(data, columns= [x[0] for x in cur.description])
```
