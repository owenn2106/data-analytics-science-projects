### Current Best Performance

```python
df['Best Performance'].value_counts().plot.pie(
    autopct="%.2f%%",
    explode=[0.25, 0],
    radius=1.25,
)
plt.show()
```

#### Based on Object dtypes

##### Job Level

```python
sns.set(rc={'figure.figsize':(7,7)})
sns.countplot(x="job_level", data=df, hue="Best Performance")
plt.title("Number of Best Peformance Employees of each Job Level")
plt.show()
```

```python
cat = 'job_level'
for i in df[cat].value_counts().index:
    print(i+": "+str((len(df.loc[(df[cat] == i) & (df['Best Performance'] == 1)]))/
          len(df.loc[df[cat] == i])))
```
